```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pymongo
import json
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["OOD_CW"]
articles_collection = db["articles"]
```

```python
articles = list(articles_collection.find({}, {"_id": 0, "content": 1, "title": 1, "author": 1}))
```

```python
for article in articles:
    content = [article['content']]
    
    # Apply TF-IDF to extract keywords for this specific article
    vectorizer = TfidfVectorizer(max_features=10, stop_words='english')
    X = vectorizer.fit_transform(content)
    
    # Extract the top 10 keywords
    keywords = vectorizer.get_feature_names_out()
    
    print(keywords)
    
    # Store the keywords back in MongoDB for this article
    articles_collection.update_one(
        {"title": article['title']},
        {"$set": {"keywords": list(keywords)}}
    )
```
