```python
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient
import datetime
```

```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client['OOD_CW']
collection = db['articles'] 
```

```python
url = 'https://techcrunch.com/'
response = requests.get(url)

# Parse the homepage HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all article links and titles
articles = soup.find_all('a', class_='loop-card__title-link')

# Extract and print titles and URLs
for article in articles:
    title = article.text.strip()  # Get the title
    content_url = article['href']  # Get the URL of the article

    print(f"Title: {title}\nURL: {content_url}\n")

print(len(articles))
```

```python
for article in articles:
    title = article.text.strip()
    content_url = article['href']
    
    # Fetch the full article page
    content_response = requests.get(content_url)
    content_soup = BeautifulSoup(content_response.text, 'html.parser')

    # Extract the article title
    article_title = content_soup.find('h1', class_='wp-block-post-title').text.strip()

    # Extract the article author (handle if author is not found)
    author_tag = content_soup.find('a', class_='post-authors-list__author')
    author = author_tag.text.strip() if author_tag else "Unknown"

    # Extract the content (you can loop through all paragraphs if needed)
    paragraphs = content_soup.find_all('p', class_='wp-block-paragraph')
    content = "\n".join([p.text for p in paragraphs])
    
    print(f"Title: {article_title}\nAuthor: {author}\nContent: {content[:100]}...\n")
    
    article_data = {
            'title': article_title,
            'author': author,
            'content': content
        }

    # Insert into MongoDB
    collection.insert_one(article_data)
    print(f"Saved article: {article_title}")
```
